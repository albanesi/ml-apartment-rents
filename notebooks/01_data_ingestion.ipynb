# 01 — Data Ingestion
Load the Inside Airbnb Zurich dataset and do a first sanity check.

```python
import pandas as pd
import os

RAW_PATH = '../data/raw/listings.csv'
PROCESSED_PATH = '../data/processed/listings_raw.csv'

df = pd.read_csv(RAW_PATH)
print(f'Shape: {df.shape}')
df.head()
```

```python
print('Columns:\n', df.columns.tolist())
```

```python
print('Missing values:\n', df.isnull().sum().sort_values(ascending=False).head(20))
```

```python
print('Dtypes:\n', df.dtypes.value_counts())
df.describe()
```

## Data Validation

```python
# Duplicate listing IDs
n_dupes = df["id"].duplicated().sum()
print(f"Duplicate IDs: {n_dupes}")

# Summary stats for key numeric columns
key_cols = ["price", "accommodates", "bedrooms", "bathrooms", "minimum_nights"]
price_raw = df["price"].replace("[\$,]", "", regex=True).astype(float)
df_check = df[key_cols[1:]].copy()
df_check.insert(0, "price", price_raw)
df_check.describe().round(2)
```

```python
# Price outlier check
p01 = price_raw.quantile(0.01)
p99 = price_raw.quantile(0.99)
n_zero = (price_raw == 0).sum()
n_low  = (price_raw < 10).sum()
n_high = (price_raw > p99).sum()
print(f"Price range: {price_raw.min():.0f} – {price_raw.max():.0f} CHF")
print(f"1st / 99th percentile: {p01:.0f} / {p99:.0f} CHF")
print(f"Zero-price listings: {n_zero}")
print(f"Price < 10 CHF: {n_low}")
print(f"Price > 99th pct ({p99:.0f} CHF): {n_high} ({100*n_high/len(price_raw):.1f}%)")
```

```python
# Train/test leakage check
# After train_test_split(test_size=0.2, random_state=42), rows must not overlap
X_train_df = pd.read_csv("../data/processed/X_train.csv")
X_test_df  = pd.read_csv("../data/processed/X_test.csv")
listings_eda = pd.read_csv("../data/processed/listings_eda.csv")

print(f"Listings after EDA cleaning:  {len(listings_eda)}")
print(f"Train rows: {len(X_train_df)} ({100*len(X_train_df)/len(listings_eda):.0f}%)")
print(f"Test rows:  {len(X_test_df)} ({100*len(X_test_df)/len(listings_eda):.0f}%)")
print(f"Train + Test = {len(X_train_df)+len(X_test_df)} — matches EDA total: {len(X_train_df)+len(X_test_df)==len(listings_eda)}")
```

```python
os.makedirs('../data/processed', exist_ok=True)
df.to_csv(PROCESSED_PATH, index=False)
print(f'Saved to {PROCESSED_PATH}')
```
